```python
import pandas as pd
```

```python
import geopandas as gpd
```

```python
cantones = gpd.read_file("datos2/cantones-crtm05.geojson")
cantones
```

```python
cantones.plot()
```

```python
redvial = gpd.read_file("datos2/redvial-crtm05.geojson")
redvial
```

```python
redvial.plot()
```

### Procesamiento
Obtención de segmentos de red vial intersecados con cantones

```python
interseccion_redvial_cantones = gpd.overlay(redvial, cantones, how='intersection')
interseccion_redvial_cantones
```

### Cálculo de longitud de segmentos de red vial intersecados con cantones

```python
interseccion_redvial_cantones["longitud_redvial"] = interseccion_redvial_cantones["geometry"].length/1000
interseccion_redvial_cantones[["categoria", "codigo", "canton", "geometry", "longitud_redvial"]]
```

#### Suma por cantón de las longitudes de los segmentos de red vial intersecados

```python
# Agrupación y suma
suma_redvial_cantones = interseccion_redvial_cantones.groupby(["cod_canton"])["longitud_redvial"].sum()

# Para convertir la serie en dataframe
suma_redvial_cantones = suma_redvial_cantones.reset_index()

suma_redvial_cantones
```

# Cálculo de la densidad de la red vial por cantón

```python
# Se le agrega a la capa de cantones la longitud de la red vial para cada polígono
cantones_redvial = cantones.merge(suma_redvial_cantones, on="cod_canton")

# Se calcula el área del cantón
cantones_redvial["area_canton"] = cantones_redvial["geometry"].area/1000000

# Se calcula la densidad
cantones_redvial["densidad_redvial"] = cantones_redvial["longitud_redvial"] / cantones_redvial["area_canton"]

cantones_redvial.sort_values(by="densidad_redvial", ascending=False)
```

##### Salidas
Mapa de coropletas de densidad de red vial en cantones

```python
cantones_redvial.plot(column="densidad_redvial",
                      legend=True,
                      cmap='Reds', 
                      edgecolor="blue",
                      scheme='quantiles',
                      figsize=(20, 20))
```
